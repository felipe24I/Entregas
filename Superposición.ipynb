```python
%matplotlib inline
#para graficar en el entorno
import sympy as sym #importo librería sympy
t=sym.symbols("t",real=True) #defino variable simbólica
#definir parametros
A=.3
om1=3
B=.5
om2=5
x=A*sym.cos(om1*t)+B*sym.cos(om2*t)
sym.plot(x,(t,-5,5),ylim=[-1.2,1.2],ylabel=r'$x(t)$')
```
