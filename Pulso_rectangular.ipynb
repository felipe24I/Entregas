```python
import sympy as sym #importo libreria sympy
t= sym.symbols("t",real=True) #defino vaable simbolica
class rect(sym.Function): #clase/objeto tipo sympy
  @classmethod #decorador objeto
  def eval(cls,arg):
    return sym.Heaviside(arg+sym.S.Half) - sym.Heaviside(arg-sym.S.Half)
sym.plot(rect(t),(t,-1,1),ylim=[-0.2,1.2],ylabel=r"Rect$(t)$")
```

```python
'''Ejercicio: Use la función rect(t) para construir una función cos(wt) con w=2 , la cual se enciende en t=-1/2  y se apaga en t=1/2'''
import sympy as sym
t,omega= sym.symbols("t omega",real=True)


class rect(sym.Function): #clase/objeto tipo sympy
  @classmethod #decorador de objeto
  def eval (cls,arg):
    return sym.Heaviside(arg+sym.S.Half)-sym.Heaviside(arg-sym.S.Half)
x=sym.cos(omega*t)*rect(t)
sym.plot(x.subs(omega,2),(t,-1,1),ylim=[-1,1],ylabel=r"$cos(wt)$")
```
