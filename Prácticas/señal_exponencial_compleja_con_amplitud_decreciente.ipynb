```python
#señal exponencial compleja con amplitud decreciente

%matplotlib inline
#Para graficar en el entorno
import sympy as sym #importar librería sympy
sym.init_printing() #inicializar para mostrar en pantalla
t,sigma,omega = sym.symbols("t sigma omega",real=True) #definir valores simbolicos
s = sigma+1j*omega #sobre cargo operador para definir parte imagniaria
x= sym.exp(s*t) #funcion exponecial simbolica desde sympy
x

y=x.subs({omega:7, sigma:-1}) #establecer valores numéricos a omega y sigma
print(r"x(t)= ",y)
sym.plot(sym.re(y),(t,0,2*sym.pi),ylabel=r"Re{$e^{st}$}",ylim=[-1,1]) #graficar parte real de la señal exponecial compleja vs el tiempo
sym.plot(sym.im(y),(t,0,2*sym.pi),ylabel=r"Im{$e^{st}$}",ylim=[-1,1]) #graficar parte imaginaria de la señal exponecial compleja vs el tiempo
```
