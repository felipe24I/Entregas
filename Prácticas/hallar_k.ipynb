```python
import sympy as sym
t= sym.symbols("t",real=True)
delta=sym.DiracDelta(2*t-5*sym.pi)
delta
x=(t**4)*sym.exp(-sym.cos(t))*sym.cos(-2*t)
sym.integrate(x*delta,(t,-sym.oo,sym.oo))
```

```python
from sympy.functions.special.delta_functions import DiracDelta
import sympy as sym
t,n,to,k,m,A=sym.symbols("t n to k m A",real=True)
step_1=sym.Heaviside(t-to)
step_2=sym.Heaviside(t-n*to)
delta=DiracDelta(t-m*to)
A=sym.integrate(step_1-step_2-(3*k)*delta,(t,-sym.oo,sym.oo))
```
