**Ejercicio**

El siguiente bloque de código permite graficar las primeras 50 muestras de la señal  x[n]=cos[2πfn]  para diferentes valores de  f .

Saque sus propias conclusiones de cómo afecta el valor de la frecuencia el periodo de la señal discreta.

```python
from numpy import exp, pi, arange, real, imag, abs, angle, sqrt
from numpy import sin, sqrt, cos
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML
import ipywidgets
def senial(f):

  n = arange(0, 50)
  plt.figure(1, (25, 10))
  x = cos(2*pi*f*n)
  plt.stem(n, x, use_line_collection=True)
  plt.grid()
  plt.margins(x=.01)
  plt.title('Gráfica de la señal $x[n] = cos(2\pi \cdot {} \cdot n)$'.format(f), fontsize=15)
  plt.xlabel('Muestras')
  plt.show()
display(ipywidgets.interactive(senial, f=ipywidgets.FloatText(value=0, step=.1, description='Valor de f:')))
```

Solución

Recordemos que en discreto, la frecuencia es cuantos ciclos se completan en n muestras.

Mientras que el periodo es cuantas muestras se tienen que tomar antes de que la señal se repita

La relación entre frecuencia y periodo esta dada por $$ f = \frac{1}{N} $$

Por lo que despejando, el periodo es

$$N = \frac{1}{f}$$

osea que si aumentamos la frecuencia, el periodo será cada vez menor, asi teniendo una relación inversamente proporcional.

Ahora, cuando f es un número entero, vamos a obtener periodicidad en la señal discreta y vamos a obtener que cada muestra va a tomar el mismo valor.

Ej: $$f=1$$

$$n=0 : x[0]= cos(2π * 0)=cos(0)=1$$

$$n=1 : x[1]= cos(2π * 1)=cos(2π)=1$$
.

.

$$n=50 : x[50]= cos(2π * 50)=cos(100π)=1$$

Vease que todas las muestras van a tomar el valor 1

$$n=0 : x[0]= cos(4π * 0)=cos(0)=1$$

$$n=1 : x[1]= cos(4π * 1)=cos(4π)=1 $$

$$n=50 : x[50]= cos(4\pi * 50)=cos(200\pi)=1$$

Vease que todas las muestras van a tomar el valor 1

$$f=0.5$$

$$n=0 : x[0]= cos(π* 0)=cos(0)=1$$

$$n=1 : x[1]= cos(π * 1)=cos(π)=-1$$

$$n=49 : x[2]= cos(π * 49)=cos(49\pi)=-1$$

$$n=50 : x[50]= cos(π * 50)=cos(50\pi)=1$$

Vease que las muestras alternan su valor entre 1 y -1

Por lo que frecuencias enteras, tienen el mismo valor porque no afectan la periodicidad, mientras que frecuencias que no son enteras afectan la periodicidad y los valores que tomen las muestras van a ser diferentes
