```python
%matplotlib inline
#para graficar en el entorno
import sympy as sym
t,a= sym.symbols("t a",real=True)
rect= sym.Heaviside(t+1/2)-sym.Heaviside(t-1/2)
x=rect.subs(t,t-1/2)+(2/3)*rect.subs(t,t-3/2)+(1/3)*rect.subs(t,t-5/2)
y=x.subs(t,a*t)

px=sym.plot(x,(t,-7,7),ylim=[-0.2,1.2],ylabel=r"$x(t)$",show=False,line_color="gray")
py=sym.plot(y.subs(a,1/2),(t,-7,7),ylim=[-0.2,1.2],ylabel=r"$y(t)$",show=False)
px.extend(py)
px.show()
```

```python
%matplotlib inline
import sympy as sym
t,a=sym.symbols("t a",real=True)
rect=sym.Heaviside(t+1/2)-sym.Heaviside(t-1/2)
x=rect.subs(t,t-1/2)+(2/3)*rect.subs(t,t-3/2)+(1/3)*rect.subs(t,t-5/2)
y=x.subs(t,a*t)

px=sym.plot(x,(t,-7,7),ylim=[-0.2,2.2],ylabel=r"$x(t)$",show=False,line_color="gray")
py=sym.plot(y.subs(a,2),(t,-7,7),ylim=[-0.2,2.2],ylabel=r"$y(t)$",show=False)
px.extend(py)
px.show()
```
