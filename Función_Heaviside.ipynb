```python
import sympy as sym #importo librería sympy
t= sym.symbols("t",real=True) #defino variable simbólica
step=sym.Heaviside(t) #defino función heaviside
step #muestro función heaviside

sym.plot(step,(t,-2,2),ylim=[-0.2,1.2],ylabel=r"$\epsilon(t)$")
```

```python
'''Ahora, se construye una función armonica (señal coseno o seno) cos(wt) con w=2 el cual se enciende en t=0
   x(t)=cos(wt)*E(t)
'''
omega= sym.symbols("omega",real=True)
x=sym.cos(omega*t)*sym.Heaviside(t)
sym.plot(x.subs(omega,2),(t,-2,10),ylim=[-1.2,1.2],ylabel=r"$\epsilon(t)$")
```
