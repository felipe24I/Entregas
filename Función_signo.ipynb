```python
import sympy as sym #importo librería sympy
t=sym.symbols("t",real=True) #defino variable simbolica
sgn=sym.sign(t) #defino función signo
sgn #muestro función signo
sym.plot(sgn,(t,-2,2),ylim=[-1.2,1.2],ylabel=r"Sgn$(t)$")
```

```python
#Genere un rectificador de onda completa utilizando la función sgn(t)
import sympy as sym #importo libreria sympy
import numpy as np
import matplotlib.pyplot as plt
t= sym.symbols("t",real=True) #defino variable simbólica
señal_entrada=sym.sin(t) #defino señal de entrada
sgn=sym.sign(sym.sin(t)) #defino señal sgn(t)
señal_rectificada=señal_entrada*sgn #calculo la señal rectificada

py=sym.plot(señal_rectificada,(t,-2,4*sym.pi),ylim=[-2,2],ylabel=r"$señal_entrada*sgn $",show=False)
px=sym.plot(señal_entrada,(t,-2,4*sym.pi),ylim=[-2,2],ylabel=r"$sin(wt)$",show=False,line_color="yellow")
px.extend(py)
px.show()

```
