```python
import sympy as sym #importo librería sympy
t=sym.symbols("t",real=True) #defino variable simbolica
sgn=sym.sign(t) #defino función signo
sgn #muestro función signo
sym.plot(sgn,(t,-2,2),ylim=[-1.2,1.2],ylabel=r"Sgn$(t)$")
```

```python
#Genere un rectificador de onda completa utilizando la función sgn(t)
import sympy as sym #importo libreria sympy
import numpy as np
import matplotlib.pyplot as plt
t= sym.symbols("t",real=True) #defino variable simbólica
señal_entrada=sym.sin(t) #defino señal de entrada
sgn=sym.sign(t) #defino señal sgn(t)
señal_rectificada=np.abs(señal_entrada*sgn) #calculo la señal rectificada


#converitr las señales sympy a señales numpy
señal_entrada_np=sym.lambdify(t,señal_entrada,'numpy')
señal_rectificada_np=sym.lambdify(t,señal_rectificada,'numpy')

#crear vector de tiempo
t_valores=np.linspace(0,4*np.pi,1000)

#calcular los valores de la señal para los valores de tiempo
señal_entrada_valores=señal_entrada_np(t_valores)
señal_rectificada_valores=señal_rectificada_np(t_valores)

#imprimir las señales
plt.plot(t_valores,señal_entrada_valores)
plt.plot(t_valores,señal_rectificada_valores)
plt.xlabel("Tiempo[Seg]")
plt.ylabel("Amplitud[V]")
plt.title("Rectificador de onda completa usando sgn(t)")
plt.grid(True)
plt.show()


```
